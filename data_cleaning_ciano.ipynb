```python
# COMPREHENSIVE DATA QUALITY ANALYSIS
print("="*60)
print("COMPREHENSIVE DATA QUALITY ANALYSIS")
print("="*60)

# 1. Basic dataset info
print(f"Dataset shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
print("\n" + "-"*50)

# 2. Check for missing values (NaN, empty strings, None)
print("MISSING VALUES ANALYSIS:")
print("\n2a. Standard missing values (NaN):")
missing_counts = df.isnull().sum()
print(missing_counts[missing_counts > 0])

print("\n2b. Empty strings:")
empty_strings = {}
for col in df.columns:
    if df[col].dtype == 'object':  # Only check string columns
        empty_count = (df[col] == '').sum()
        if empty_count > 0:
            empty_strings[col] = empty_count
if empty_strings:
    for col, count in empty_strings.items():
        print(f"{col}: {count} empty strings")
else:
    print("No empty strings found")

# 3. Check for problematic values
print("\n" + "-"*50)
print("PROBLEMATIC VALUES ANALYSIS:")

# Check each column for issues
for col in df.columns:
    print(f"\n3. Column '{col}':")
    unique_values = df[col].value_counts()
    
    # Look for suspicious values
    suspicious = []
    if col in ['Item', 'Payment Method', 'Location']:
        # Check for UNKNOWN, ERROR, NaN strings
        for val in unique_values.index:
            if pd.isna(val) or str(val).upper() in ['UNKNOWN', 'ERROR', 'NAN']:
                suspicious.append((val, unique_values[val]))
    
    elif col in ['Price Per Unit', 'Quantity', 'Total Spent']:
        # Check for non-numeric or negative values
        for val in unique_values.index:
            if pd.isna(val) or str(val).upper() in ['ERROR', 'NAN'] or (isinstance(val, (int, float)) and val < 0):
                suspicious.append((val, unique_values[val]))
    
    if suspicious:
        print(f"   Issues found: {suspicious}")
    else:
        print(f"   No obvious issues (showing top 5 values)")
        print(f"   {unique_values.head().to_dict()}")

print("\n" + "-"*50)
print("DATA TYPE ANALYSIS:")
print(df.dtypes)

print("\n" + "-"*50)
print("SAMPLE OF POTENTIALLY PROBLEMATIC ROWS:")
# Show rows with various issues
problem_mask = (
    (df['Item'].isna()) | 
    (df['Item'] == '') |
    (df['Payment Method'].str.contains('ERROR|UNKNOWN', na=False)) |
    (df['Location'].str.contains('UNKNOWN', na=False)) |
    (df['Total Spent'].astype(str).str.contains('ERROR', na=False))
)

problem_rows = df[problem_mask]
print(f"Found {len(problem_rows)} problematic rows:")
if len(problem_rows) > 0:
    print(problem_rows.head(10))
```

```python
# COMPREHENSIVE DATA CLEANING IMPLEMENTATION
print("="*60)
print("IMPLEMENTING DATA CLEANING STEPS")
print("="*60)

# Create a copy for cleaning
df_clean = df.copy()
print(f"Starting with {df_clean.shape[0]} rows")

# STEP 1: Handle missing and empty values
print("\nSTEP 1: Handling missing and empty values...")

# 1a. Remove rows where Item is missing or empty
initial_count = len(df_clean)
df_clean = df_clean.dropna(subset=['Item'])  # Remove NaN in Item
df_clean = df_clean[df_clean['Item'] != '']  # Remove empty strings in Item
print(f"   Removed {initial_count - len(df_clean)} rows with missing/empty Items")

# 1b. Fill missing Payment Methods and Locations
df_clean['Payment Method'] = df_clean['Payment Method'].fillna('Unknown')
df_clean['Location'] = df_clean['Location'].fillna('Unknown')
print("   Filled missing Payment Methods and Locations with 'Unknown'")

# STEP 2: Handle problematic values
print("\nSTEP 2: Cleaning problematic values...")

# 2a. Fix 'UNKNOWN' values (replace with 'Unknown' for consistency)
df_clean['Payment Method'] = df_clean['Payment Method'].replace('UNKNOWN', 'Unknown')
df_clean['Location'] = df_clean['Location'].replace('UNKNOWN', 'Unknown')
print("   Standardized 'UNKNOWN' to 'Unknown'")

# 2b. Handle 'ERROR' values in Payment Method
error_payment_count = (df_clean['Payment Method'] == 'ERROR').sum()
df_clean['Payment Method'] = df_clean['Payment Method'].replace('ERROR', 'Unknown')
print(f"   Fixed {error_payment_count} 'ERROR' values in Payment Method")

# STEP 3: Data type conversions (DO THIS BEFORE mathematical operations)
print("\nSTEP 3: Converting data types...")

# Convert numeric columns to proper types first
numeric_columns = ['Quantity', 'Price Per Unit', 'Total Spent']
for col in numeric_columns:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

print("   Converted numeric columns to proper types")

# 2c. NOW handle 'ERROR' values in Total Spent (after conversion)
error_total_mask = df_clean['Total Spent'].isna()  # After conversion, 'ERROR' becomes NaN
error_total_count = error_total_mask.sum()
if error_total_count > 0:
    # Calculate correct total: Quantity * Price Per Unit
    df_clean.loc[error_total_mask, 'Total Spent'] = (
        df_clean.loc[error_total_mask, 'Quantity'] * 
        df_clean.loc[error_total_mask, 'Price Per Unit']
    )
    print(f"   Fixed {error_total_count} 'ERROR' values in Total Spent by recalculating")

# Convert Transaction Date to datetime
df_clean['Transaction Date'] = pd.to_datetime(df_clean['Transaction Date'], errors='coerce')

# STEP 4: Remove invalid/impossible values
print("\nSTEP 4: Removing invalid values...")

# 4a. Remove rows with negative or zero quantities/prices/totals or NaN values
initial_count = len(df_clean)
df_clean = df_clean.dropna(subset=['Quantity', 'Price Per Unit', 'Total Spent'])
df_clean = df_clean[
    (df_clean['Quantity'] > 0) & 
    (df_clean['Price Per Unit'] > 0) & 
    (df_clean['Total Spent'] > 0)
]
print(f"   Removed {initial_count - len(df_clean)} rows with invalid/missing quantities/prices")

# 4b. Remove rows where Total Spent doesn't match Quantity * Price (allowing small rounding errors)
tolerance = 0.01
calculated_total = df_clean['Quantity'] * df_clean['Price Per Unit']
mismatch_mask = abs(df_clean['Total Spent'] - calculated_total) > tolerance
mismatch_count = mismatch_mask.sum()
df_clean = df_clean[~mismatch_mask]
print(f"   Removed {mismatch_count} rows with incorrect total calculations")

# STEP 5: Final cleanup
print("\nSTEP 5: Final cleanup...")

# Remove any remaining rows with NaN in critical columns
df_clean = df_clean.dropna(subset=['Item', 'Transaction Date'])

# Reset index
df_clean = df_clean.reset_index(drop=True)

print(f"\nCLEANING COMPLETE!")
print(f"Original dataset: {df.shape[0]} rows")
print(f"Cleaned dataset: {df_clean.shape[0]} rows")
print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")
print(f"Data retention rate: {df_clean.shape[0]/df.shape[0]*100:.1f}%")

# Show sample of cleaned data
print(f"\nSample of cleaned data:")
df_clean.head()
```

```python
# Check how many rows were affected by the cleaning process
print("ROWS AFFECTED BY CLEANING:")
print("="*40)

original_rows = df.shape[0]
cleaned_rows = df_clean.shape[0]
rows_removed = original_rows - cleaned_rows
retention_rate = (cleaned_rows / original_rows) * 100

print(f"Original dataset: {original_rows:,} rows")
print(f"Cleaned dataset:  {cleaned_rows:,} rows")
print(f"Rows removed:     {rows_removed:,} rows")
print(f"Retention rate:   {retention_rate:.1f}%")
print(f"Removal rate:     {100-retention_rate:.1f}%")
```

```python
# SAVE CLEANED DATA AND UPDATE ORIGINAL
print("="*50)
print("SAVING CLEANED DATA")
print("="*50)

# Option 1: Save as a new CSV file (recommended)
df_clean.to_csv('cafe_sales_cleaned.csv', index=False)
print("✅ Saved cleaned data as 'cafe_sales_cleaned.csv'")

# Option 2: Create backup of original and overwrite (if you want to replace original)
import os
if os.path.exists('cafe_sales.csv'):
    # Create backup first
    df.to_csv('cafe_sales_backup.csv', index=False)
    print("✅ Created backup as 'cafe_sales_backup.csv'")
    
    # Overwrite original file with cleaned data
    df_clean.to_csv('cafe_sales.csv', index=False)
    print("✅ Overwritten 'cafe_sales.csv' with cleaned data")

# Option 3: Update the original DataFrame variable in memory
df = df_clean.copy()
print("✅ Updated original 'df' variable with cleaned data")

print(f"\nFinal Summary:")
print(f"- Original file backed up: cafe_sales_backup.csv")
print(f"- Cleaned file saved: cafe_sales_cleaned.csv") 
print(f"- Original file updated: cafe_sales.csv")
print(f"- DataFrame 'df' now contains {df.shape[0]} clean rows")

# Verify the save worked
print(f"\nVerification:")
print(f"df shape: {df.shape}")
print(f"df_clean shape: {df_clean.shape}")
print(f"Are they identical? {df.equals(df_clean)}")
```
